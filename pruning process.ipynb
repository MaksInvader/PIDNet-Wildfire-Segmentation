```python
def parse_event(p):
    parts = p.stem.split("_")
    if len(parts) < 3:
        return "Unknown"

    # Assuming Date is parts[0] and Frame is parts[-1]
    # We take everything in the middle as the 'Event'
    event = "_".join(parts[1:-1])
    return event
```

```python
import os

# --- CONFIGURATION ---
image_dir = '/content/Labeled HPWREN/fire_images_merged'
label_dir = '/content/labels'
# ---------------------

def verify_dataset():
    # Get all base names (filename without extension)
    image_files = {os.path.splitext(f)[0]: f for f in os.listdir(image_dir)
                   if f.lower().endswith(('.jpg', '.jpeg', '.png'))}
    label_files = {os.path.splitext(f)[0]: f for f in os.listdir(label_dir)
                   if f.lower().endswith('.txt')}

    image_set = set(image_files.keys())
    label_set = set(label_files.keys())

    # 1. Check for Images missing Labels
    missing_labels = image_set - label_set

    # 2. Check for Labels missing Images (Orphan labels)
    orphan_labels = label_set - image_set

    # 3. Check for Empty Labels (Potential Class 0 issues)
    empty_labels = []
    class_counts = {}

    for name, filename in label_files.items():
        path = os.path.join(label_dir, filename)
        if os.path.exists(path):
            with open(path, 'r') as f:
                lines = [line.strip() for line in f.readlines() if line.strip()]
                if not lines:
                    empty_labels.append(filename)
                else:
                    for line in lines:
                        cls = line.split()[0]
                        class_counts[cls] = class_counts.get(cls, 0) + 1

    # --- REPORTING ---
    print("--- DATASET INTEGRITY REPORT ---")
    print(f"Total Images Found: {len(image_set)}")
    print(f"Total Labels Found: {len(label_set)}")
    print("-" * 30)

    if not missing_labels and not orphan_labels:
        print("✅ 100% MATCH: Every image has a corresponding label file.")
    else:
        if missing_labels:
            print(f"❌ MISSING LABELS: {len(missing_labels)} images have no .txt file.")
            print(f"   Example missing: {list(missing_labels)[:3]}")
        if orphan_labels:
            print(f"⚠️ ORPHAN LABELS: {len(orphan_labels)} .txt files have no image.")

    print("-" * 30)
    print(f"Empty Label Files: {len(empty_labels)}")
    print("Class Distribution (instances):")
    for cls, count in sorted(class_counts.items()):
        print(f"  Class {cls}: {count}")

    return missing_labels

# Run the verification
missing = verify_dataset()
```

```python
import os
import random
import shutil
from collections import defaultdict

# --- CONFIGURATION ---
image_dir = '/content/Labeled HPWREN/fire_images_merged'
label_dir = '/content/labels'
output_dir = '/content/Labeled HPWREN/pruned_dataset_safe'

BG_TARGET_TOTAL = 2500
C1_MAX_PER_GROUP = 50
# ---------------------

def run_pruning():
    # Ensure output directories exist
    os.makedirs(os.path.join(output_dir, 'images'), exist_ok=True)
    os.makedirs(os.path.join(output_dir, 'labels'), exist_ok=True)

    class_map = {}
    groups = defaultdict(list)

    # 1. Indexing and Class Mapping
    print("Indexing files and reading labels...")
    files = [f for f in os.listdir(image_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

    for filename in files:
        # Grouping Logic (Middle string)
        parts = filename.split('_')
        if len(parts) >= 3:
            middle_string = "_".join(parts[1:-1])
            groups[middle_string].append(filename)
        else:
            # Fallback for files that don't follow the underscore convention
            groups['ungrouped'].append(filename)

        # Label Parsing
        base_name = os.path.splitext(filename)[0]
        label_path = os.path.join(label_dir, f"{base_name}.txt")

        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                lines = [line.strip() for line in f.readlines() if line.strip()]
                if not lines:
                    class_map[filename] = -1  # Explicitly mark as Background (Empty Label)
                else:
                    # Collect all classes in the file
                    classes = [int(line.split()[0]) for line in lines]
                    if 0 in classes:
                        class_map[filename] = 0   # Priority 1: Fire
                    elif 1 in classes:
                        class_map[filename] = 1   # Priority 2: Smoke
                    else:
                        class_map[filename] = classes[0] # Other
        else:
            class_map[filename] = -1 # Background (Missing Label)

    # 2. Tiered Pruning Logic
    selected_files = set() # Using a set to prevent any duplicates
    print(f"Processing {len(groups)} groups...")

    for middle, group_files in groups.items():
        group_files.sort() # Keep chronological order

        # Check the whole group for class presence
        group_classes = [class_map.get(f, -1) for f in group_files]

        if 0 in group_classes:
            # TIER 1: Sequence contains Fire. KEEP ENTIRE SEQUENCE.
            for f in group_files:
                selected_files.add(f)

        elif 1 in group_classes:
            # TIER 2: Sequence contains Smoke (no Fire). Downsample.
            step = max(1, len(group_files) // C1_MAX_PER_GROUP)
            sampled = group_files[::step]
            for f in sampled:
                selected_files.add(f)

        else:
            # TIER 3: Pure Background sequence. Sparse sampling.
            step = max(1, len(group_files) // 10) # Initial reduction
            sampled = group_files[::step]
            for f in sampled:
                selected_files.add(f)

    # 3. SAFETY NET: Explicitly ensure ALL Class 0 images from original dataset are included
    all_c0_original = [f for f, cls in class_map.items() if cls == 0]
    for f in all_c0_original:
        selected_files.add(f)

    # 4. Final Background Hard-Cap
    # Separate the current selection into "Must-Keep" and "Background"
    must_keep = [f for f in selected_files if class_map.get(f, -1) != -1]
    background_pool = [f for f in selected_files if class_map.get(f, -1) == -1]

    if len(background_pool) > BG_TARGET_TOTAL:
        print(f"Trimming background images from {len(background_pool)} down to {BG_TARGET_TOTAL}...")
        final_bg = random.sample(background_pool, BG_TARGET_TOTAL)
        final_selection = must_keep + final_bg
    else:
        final_selection = list(selected_files)

    # 5. Copying Files
    print(f"Copying {len(final_selection)} selected files...")
    for f in final_selection:
        # Copy image
        shutil.copy2(os.path.join(image_dir, f), os.path.join(output_dir, 'images', f))

        # Copy label
        base_name = os.path.splitext(f)[0]
        label_f = f"{base_name}.txt"
        shutil.copy2(os.path.join(label_dir, label_f), os.path.join(output_dir, 'labels', label_f))

    # 6. VERIFICATION REPORT
    final_images = os.listdir(os.path.join(output_dir, 'images'))
    c0_count = 0
    c1_count = 0
    bg_count = 0

    for f in final_images:
        cls = class_map.get(f)
        if cls == 0: c0_count += 1
        elif cls == 1: c1_count += 1
        elif cls == -1: bg_count += 1

    print("\n" + "="*30)
    print("FINAL DATASET REPORT")
    print(f"Total Images: {len(final_images)}")
    print(f"Class 0 (Fire): {c0_count}  <-- Should be at least original 1904")
    print(f"Class 1 (Smoke): {c1_count}")
    print(f"Background: {bg_count}")
    print("="*30)

if __name__ == "__main__":
    run_pruning()
```

```python
import os
import random
import shutil
from collections import defaultdict

# --- CONFIGURATION ---
image_dir = '/content/Labeled HPWREN/fire_images_merged'
label_dir = '/content/labels' # Adjusted based on your path
output_dir = '/content/Labeled HPWREN/pruned_datasetreal'

BG_TARGET_TOTAL = 2500
C1_MAX_PER_GROUP = 50
# ---------------------

os.makedirs(os.path.join(output_dir, 'images'), exist_ok=True)
os.makedirs(os.path.join(output_dir, 'labels'), exist_ok=True)

class_map = {}
groups = defaultdict(list)

# 1. Map labels and group images using your underscore logic
print("Indexing files and reading labels...")
files = [f for f in os.listdir(image_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

for filename in files:
    parts = filename.split('_')

    # Apply your middle string logic
    if len(parts) >= 3:
        middle_string = "_".join(parts[1:-1])
        groups[middle_string].append(filename)

        # Look for the corresponding label
        base_name = os.path.splitext(filename)[0]
        label_path = os.path.join(label_dir, f"{base_name}.txt")

        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                lines = f.readlines()
                # Extract first integer (class) from each line
                classes = [int(line.split()[0]) for line in lines if line.strip()]

                # Priority mapping
                if 0 in classes:
                    class_map[filename] = 0
                elif 1 in classes:
                    class_map[filename] = 1
                else:
                    class_map[filename] = classes[0] if classes else None
        else:
            class_map[filename] = None

# 2. Apply the 3-Tier Pruning Logic
selected_files = []
print(f"Processing {len(groups)} groups...")

for middle, group_files in groups.items():
    group_files.sort() # Keep chronological order
    group_classes = [class_map.get(f) for f in group_files]

    if 0 in group_classes:
        # TIER 1: Keep everything in this sequence
        selected_files.extend(group_files)
    elif 1 in group_classes:
        # TIER 2: Sample to reduce redundancy
        step = max(1, len(group_files) // C1_MAX_PER_GROUP)
        selected_files.extend(group_files[::step])
    else:
        # TIER 3: Sparse background sampling
        step = max(1, len(group_files) // 5)
        selected_files.extend(group_files[::step])

# 3. Final Background Hard-Cap
bg_only = [f for f in selected_files if class_map.get(f) is None]
if len(bg_only) > BG_TARGET_TOTAL:
    print(f"Trimming background images from {len(bg_only)} down to {BG_TARGET_TOTAL}...")
    final_bg = random.sample(bg_only, BG_TARGET_TOTAL)
    # Rebuild list: Keep all with classes + the capped backgrounds
    selected_files = [f for f in selected_files if class_map.get(f) is not None] + final_bg

# 4. Copy to New Destination
print(f"Copying {len(selected_files)} selected files to output folder...")
for f in selected_files:
    # Copy image
    shutil.copy2(os.path.join(image_dir, f), os.path.join(output_dir, 'images', f))

    # Copy label if it exists
    base_name = os.path.splitext(f)[0]
    label_f = f"{base_name}.txt"
    source_label = os.path.join(label_dir, label_f)
    if os.path.exists(source_label):
        shutil.copy2(source_label, os.path.join(output_dir, 'labels', label_f))

print(f"Success! Pruned dataset is ready at: {output_dir}")
```

```python
import os
import random
import shutil
from collections import defaultdict

# --- CONFIGURATION ---
image_dir = '/content/Labeled HPWREN/fire_images_merged'
label_dir = '/content/labels'
output_dir = '/content/Labeled HPWREN/pruned_dataset_safereal'

BG_TARGET_TOTAL = 2500
C1_MAX_PER_GROUP = 50
# ---------------------

def run_pruning():
    # Ensure output directories exist
    os.makedirs(os.path.join(output_dir, 'images'), exist_ok=True)
    os.makedirs(os.path.join(output_dir, 'labels'), exist_ok=True)

    image_to_classes = {}
    groups = defaultdict(list)

    # 1. Indexing and Robust Label Mapping
    print("Indexing files and performing deep label scan...")

    label_files_map = {}
    for f in os.listdir(label_dir):
        if f.lower().endswith('.txt'):
            base = os.path.splitext(f)[0].strip().lower()
            label_files_map[base] = f

    image_files = [f for f in os.listdir(image_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

    total_c0_instances = 0 # To track total boxes (like your verification script)

    for filename in image_files:
        # Grouping Logic
        parts = filename.split('_')
        if len(parts) >= 3:
            middle_string = "_".join(parts[1:-1])
            groups[middle_string].append(filename)
        else:
            groups['ungrouped'].append(filename)

        # Robust Label Parsing
        base_name = os.path.splitext(filename)[0].strip().lower()
        current_classes = set()

        if base_name in label_files_map:
            label_path = os.path.join(label_dir, label_files_map[base_name])
            try:
                with open(label_path, 'r', encoding='utf-8', errors='ignore') as f:
                    lines = f.read().strip().splitlines()
                    if not lines:
                        current_classes.add(-1)
                    else:
                        for line in lines:
                            parts = line.split()
                            if parts:
                                try:
                                    class_id = int(parts[0])
                                    current_classes.add(class_id)
                                    if class_id == 0:
                                        total_c0_instances += 1
                                except ValueError:
                                    continue
            except Exception as e:
                current_classes.add(-1)
        else:
            current_classes.add(-1)

        image_to_classes[filename] = current_classes

    # Pre-identify Class 0
    all_original_c0_files = [f for f, cls_set in image_to_classes.items() if 0 in cls_set]

    print("-" * 30)
    print(f"SCAN RESULT:")
    print(f"Total Class 0 Boxes (Instances): {total_c0_instances}")
    print(f"Total Unique Images containing Class 0: {len(all_original_c0_files)}")
    print("-" * 30)

    # 2. Tiered Pruning Logic
    selected_files = set()
    print(f"Processing {len(groups)} groups...")

    for middle, group_files in groups.items():
        group_files.sort()
        group_has_fire = any(0 in image_to_classes.get(f, set()) for f in group_files)
        group_has_smoke = any(1 in image_to_classes.get(f, set()) for f in group_files)

        if group_has_fire:
            for f in group_files:
                selected_files.add(f)
        elif group_has_smoke:
            step = max(1, len(group_files) // C1_MAX_PER_GROUP)
            sampled = group_files[::step]
            for f in sampled:
                selected_files.add(f)
        else:
            step = max(1, len(group_files) // 10)
            sampled = group_files[::step]
            for f in sampled:
                selected_files.add(f)

    # 3. Consolidation & Background Cap
    intermediate_list = list(selected_files)
    must_keep = [f for f in intermediate_list if any(c >= 0 for c in image_to_classes.get(f, set()))]
    background_pool = [f for f in intermediate_list if -1 in image_to_classes.get(f, set()) and not any(c >= 0 for c in image_to_classes.get(f, set()))]

    if len(background_pool) > BG_TARGET_TOTAL:
        print(f"Trimming background images from {len(background_pool)} down to {BG_TARGET_TOTAL}...")
        final_bg = random.sample(background_pool, BG_TARGET_TOTAL)
        final_selection = must_keep + final_bg
    else:
        final_selection = intermediate_list

    # 4. ABSOLUTE SAFETY OVERRIDE
    final_selection_set = set(final_selection)
    for f in all_original_c0_files:
        final_selection_set.add(f)

    final_selection = list(final_selection_set)

    # 5. Copying Files
    print(f"Copying {len(final_selection)} selected files to {output_dir}...")
    for f in final_selection:
        shutil.copy2(os.path.join(image_dir, f), os.path.join(output_dir, 'images', f))
        base_name = os.path.splitext(f)[0]
        base_name_lower = base_name.strip().lower()
        if base_name_lower in label_files_map:
            actual_label_name = label_files_map[base_name_lower]
            shutil.copy2(os.path.join(label_dir, actual_label_name), os.path.join(output_dir, 'labels', f"{base_name}.txt"))
        else:
            with open(os.path.join(output_dir, 'labels', f"{base_name}.txt"), 'w') as empty_f:
                pass

    # 6. VERIFICATION REPORT
    c0_final_images = 0
    c0_final_instances = 0
    for f in final_selection:
        cls_set = image_to_classes.get(f, set())
        if 0 in cls_set:
            c0_final_images += 1
            # Re-read file to count instances for the report
            base_name = os.path.splitext(f)[0].strip().lower()
            if base_name in label_files_map:
                with open(os.path.join(label_dir, label_files_map[base_name]), 'r') as lf:
                    c0_final_instances += sum(1 for line in lf if line.strip().startswith('0'))

    print("\n" + "="*30)
    print("FINAL DATASET REPORT")
    print(f"Total Images: {len(final_selection)}")
    print(f"Class 0 Images: {c0_final_images} (Target Files: {len(all_original_c0_files)})")
    print(f"Class 0 Instances: {c0_final_instances} (Target Boxes: {total_c0_instances})")
    print("="*30)

if __name__ == "__main__":
    run_pruning()
```
